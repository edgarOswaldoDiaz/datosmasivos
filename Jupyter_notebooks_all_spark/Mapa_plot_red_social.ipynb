```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
import contextily as cx
```

```python
df = gpd.read_file('tuits_georeferenciados_2021.csv')
df.crs = 'epsg:4326'
```

```python
df['geometry'] = df['PUNTO'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(df, crs='epsg:4326')
gdf.head()
```

```python
gdf.plot(column = 'PUNTO', cmap='Blues', figsize=(8, 5));
```

```python
ax = gdf.plot(color="red", markersize=10, figsize=(15, 15))
cx.add_basemap(ax, crs=gdf.crs.to_string())
```
