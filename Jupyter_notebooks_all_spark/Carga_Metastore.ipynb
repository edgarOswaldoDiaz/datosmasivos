# Carga de Metadatos

En esta sección se muestra un ejemplo para cargar metadatos en Apache Hive a traves de Trino


### Librerias python

Importamos algunas librerias extra que nos ayudan a interactuar con los datos

```python
import trino # Librerias para hacer una conexion JDBC a trino
import pandas as pd # Librerias para presentación de información en formato de tabla
```

### Definición de la conexión

Definimos la conexión a trino. El catalogo y esquema son opcionales 

```python
conn=trino.dbapi.connect(
    host='trino', #URL o IP del servicio de trino
    port=8080, # puerdo donde recibe conexiones el servico Trino
    user='jupyter', # Usuario el cual consumirá la informacion
    #catalog='warehouse', # Catalogo que se desea consultar 
    #schema='lad_enlace24', # Esquema al cual se quiere acceder 
)
```

## Funciones de ayuda

```python
# Ejecuta la sentencia sql y regresa el resultado, utilizado para creacion de tablas y vistas
def ejecutar_sql(consulta):
    cur = conn.cursor()
    cur.execute(consulta)
    return cur.fetchall()

# Ejecuta la sentencia sql y trae el resultado en forma de tabla de pandas. 
# Se utiliza mayormente en sentancias select
def sql_tabla(consulta):
    cur = conn.cursor()
    cur.execute(consulta)
    valores = cur.fetchall()
    col_names = [x[0] for x in cur.description]
    return pd.DataFrame(valores,columns=col_names)
```

```python
ejecutar_sql(consulta_crea_tabla)
```
