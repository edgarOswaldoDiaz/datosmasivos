```r
# Cargar conjunto de datos a visualizar basado en funciones de R 
tipo <- c("A","A","A","A","A",
          "B","B","B","B","B",
          "C","C","C","C","C")
anio <- c(2000,2001,2002,2003,2004,
           2000,2001,2002,2003,2004,
           2000,2001,2002,2003,2004)
set.seed(1234)
valor <- rnorm(15,25,10)
datos <- data.frame(tipo, anio, valor)
library(ggplot2)
ggplot(datos, aes(x=anio, y=valor, group = tipo, colour =tipo )) + 
  geom_line()  + 
  geom_point( size=4, shape=15, fill="red") + 
  theme_minimal()
```
