#### Recolecta información externa y genera muestra de medidas del cuerpo 

```r
url <- 'https://raw.githubusercontent.com/fhernanb/datos/master/medidas_cuerpo'
datos <- read.table(file=url, header=T)
require(MASS)  # Se debe cargar este paquete
f1 <- kde2d(x=datos$peso, y=datos$altura, n=10)
f2 <- kde2d(x=datos$peso, y=datos$altura, n=50)
par(mfrow=c(1, 2))
image(f1, xlab='Peso', ylab='Estatura', main='Muestra(a)=10')
image(f2, xlab='Peso', ylab='Estatura', main='Muestra(b)=50')
```
